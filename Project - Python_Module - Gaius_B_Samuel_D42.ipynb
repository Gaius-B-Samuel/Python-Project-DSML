## __Overview__ <br>

<span style="color:green;">●</span>  This project involves analyzing employee data from ABC Company, consisting of 458 rows and 9 columns. <br>
<span style="color:green;">●</span>  The analysis aims to extract meaningful insights about the workforce, including team distribution, salary expenditure, and age demographics. <br>
<span style="color:green;">●</span>  The project includes data preprocessing, analysis, visualization, and reporting.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
dataset_url="https://docs.google.com/spreadsheets/d/1VP9BE_eI2yl6uUHSm4mGiiwjRdoqCqnkcIjsv5Q2ex4/export?format=csv"
df=pd.read_csv(dataset_url)
```

```python
df
```

```python
# Preprocessing: Correcting the "height" column
df['height']=np.random.randint(150,181,size=len(df))
```

### Analysis Task 1: Distribution of employees across teams

```python
team_distribution=df['Team'].value_counts()
total_employees=len(df)
team_percentage=(team_distribution/total_employees)*100
```

```python
# Visualization for Task 1
plt.figure(figsize=(8,5))
team_distribution.plot(kind='bar',color='skyblue')
plt.xlabel("Teams")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Teams")
plt.show()
```

### Analysis Task 2: Segregate employees based on position

```python
position_distribution=df['Position'].value_counts()
```

```python
# Visualization for Task 2
plt.figure(figsize=(8,5))
position_distribution.plot(kind='bar',color='coral')
plt.xlabel("Positions")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution by Position")
plt.show()
```

### Analysis Task 3: Identify the predominant age group

```python
age_bins=[20, 30, 40, 50, 60]
age_labels=["20-30","30-40","40-50","50-60"]
df['age_group']=pd.cut(df['Age'],bins=age_bins,labels=age_labels)
age_group_distribution=df['age_group'].value_counts()
```

```python
# Visualization for Task 3
plt.figure(figsize=(8,5))
age_group_distribution.plot(kind='bar',color='green')
plt.xlabel("Age Groups")
plt.ylabel("Number of Employees")
plt.title("Predominant Age Group of Employees")
plt.show()
```

### Analysis Task 4: Identify team and position with highest salary expenditure

```python
team_salary=df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary=df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
```

```python
# Visualization for Task 4
plt.figure(figsize=(8,5))
team_salary.plot(kind='bar',color='purple')
plt.xlabel("Teams")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Team")
plt.show()

plt.figure(figsize=(8,5))
position_salary.plot(kind='bar',color='orange')
plt.xlabel("Positions")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Position")
plt.show()
```

### Analysis Task 5: Correlation between age and salary

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Age'], y=df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation Between Age and Salary")
plt.show()
```

```python
# Saving the processed dataset
df.to_csv("processed_employee_data.csv",index=False)
```
